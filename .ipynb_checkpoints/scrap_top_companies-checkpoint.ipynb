## Logging into indeed.com
The code below details the functions that allow us to log into indeed.com and click into it's company review site, where we can continuously search for companies we are interested in and scrap their reviews.
Note that this sequence of code will not always work, as indeed.com is able to pick up scrapping activities and have additional steps for us to authenticate that we are human.

```python
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

username_key = "abc@hotmail.com"
password = "password"

# Find the username input field and enter your email
username_input = driver.find_element_by_name("__email")
username_input.send_keys(username_key)

#Accept cookies
accept_cookies_button = WebDriverWait(driver, 10).until(
    EC.element_to_be_clickable((By.CSS_SELECTOR, "[id='onetrust-accept-btn-handler']"))
)
accept_cookies_button.click()

# Click "Use this email" button
cont_email_button = WebDriverWait(driver, 10).until(
    EC.element_to_be_clickable((By.CSS_SELECTOR, "[data-tn-element='auth-page-email-submit-button']"))
)
cont_email_button.click()

# Wait until the password input field is visible
password_input = WebDriverWait(driver, 10).until(
    EC.visibility_of_element_located((By.CSS_SELECTOR, "[name='__password']"))
)

# Enter your password
password_input.send_keys(password)

# Wait until the "Sign In" button is clickable
sign_in_button = WebDriverWait(driver, 10).until(
    EC.element_to_be_clickable((By.CSS_SELECTOR, "[data-tn-element='auth-page-sign-in-password-form-submit-button']"))
)

# Submit and login
sign_in_button.click()

#driver.get("https://sg.indeed.com/cmp/Standard-Chartered-Bank/reviews")

#click to go to company review page
company_review_button = WebDriverWait(driver, 10).until(
    EC.element_to_be_clickable((By.CSS_SELECTOR, "[id='CompanyReviews']"))
)
company_review_button.click()
```

## Functions to scrap company reviews
The process goes like this. 
1. Type the company name into the search bar to search for the company
2. Identify the most relevant result in the results page and click into their review site
3. Extract the review contents before clicking to the company review search bar in step 1

Note that for review information in the format: role (current_status) - location - date, not all information can be  extracted using this pattern. This is because, while the format mentioned is fixed, people may add extra "-" and "()" to supplement more details of their role, exact location or current status.

```python
#Search for company in search bar
def search_company(company_name):
    company_search_box = WebDriverWait(driver, 10).until(
    EC.visibility_of_element_located((By.CSS_SELECTOR, "[data-testid='company-search-box']")))
    company_search_box.send_keys(company_name)
    search_button = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.CSS_SELECTOR, "[type='submit']")))
    search_button.click()
    return 

#go into review page directly
def click_most_relevant_result():
    parent_element = WebDriverWait(driver, 10).until(
    EC.visibility_of_element_located((By.CSS_SELECTOR, "[data-tn-component='CompanyRow']")))
    child_elements = parent_element.find_elements_by_css_selector('[href]')
    filtered_elements = [element for element in child_elements if "reviews" in element.get_attribute('href')]
    
    #get industry first before clicking
    industries = WebDriverWait(driver, 10).until(
    EC.visibility_of_element_located((By.CSS_SELECTOR, "[class='css-kbqft4 e1wnkr790']")))
    industry = industries.text
    
    most_relevant = filtered_elements[0]
    most_relevant.click()
    #scrap industry

    return industry

#scrap review page 
def scrap_review_page(company,industry):
    #focus only on review list
    stop = False
    while stop ==False:
        reviewlist = WebDriverWait(driver, 10).until(
        EC.visibility_of_element_located((By.CSS_SELECTOR, "[class = 'cmp-ReviewsList']")))
        #search review by review all review are under class = css-lw17hn eu4oa1w0
        reviews = reviewlist.find_elements_by_css_selector("[class = 'css-lw17hn eu4oa1w0']")
        
        #Extract each individual review
        for review in reviews:
            try:
                review.find_element_by_css_selector('span[itemprop="author"]')
                text = review.text
                #Remove uncessary text in featured reviews
                if "Indeed featured review\n" in text:
                    text=text.replace("Indeed featured review\n","")
                    text=text.replace("The most useful review selected by Indeed","")    
                    
                info_list = text.split('\n')
                user_info = info_list[2]
                #the information of the review are in this format Role (current status) - Location - date
                user_info_1=user_info.split("-")[0].strip()
                user_info_2=user_info.split("-")[1:]
                role = user_info_1
                location = user_info_2[0].strip()
                date = user_info_2[1].strip()


                #extract review write up
                review_writeup =" ".join(info_list[3:-5]).replace("Was this review helpful?","")
                review_dict["review"].append(review_writeup)


                #extract main ratings
                parent = review.find_element_by_css_selector('div[itemprop="reviewRating"]')
                main_rating = parent.find_element_by_tag_name('button').text
                review_dict["main_rating"].append(main_rating)

                # Extract the sub ratings
                sub_ratings = parent.find_elements_by_css_selector('.css-1exgkxc.e37uo190')

                # Iterate through the sub ratings and extract the information
                sub_ratings_info = []
                for sub_rating,cat in zip(sub_ratings,["work_life_balance_rating","salary_and_benefits_rating",
                                                            "job_security_and_advancement_rating",
                                                           "management_rating", "job_culture_rating"]):
                    rating = sub_rating.find_element_by_css_selector("[role='img']").get_attribute('aria-label')
                    rating = rating.split(" ")[0]
                    review_dict[cat].append(rating)
            #append everything at the end, so that if there is an error i can skip that review
                review_dict["company"].append(company)
                review_dict["industry"].append(industry)
                review_dict["role"].append(role)
                review_dict["location"].append(location)
                review_dict["date"].append(date)
            except:
                pass
        #go to the next page of reviews if it exist
        try:
            next_button = WebDriverWait(driver, 10).until(
                EC.element_to_be_clickable((By.CSS_SELECTOR, "[data-tn-element='next-page']")))
            next_button.click()
        except:
            stop = True
            
#Check whether singapore is a valid location in the location filter list, if it isn't we will not be extracting the data as 
#we are only extracting reviews from Singapore.
def check_location():
    parent = WebDriverWait(driver, 10).until(
        EC.visibility_of_all_elements_located((By.CSS_SELECTOR, "[data-testid = 'selected-value']")))
    if parent[1].text != "Singapore": #index 1 for location bar
        return False
    else:
        return True
```

## Running the functions to scrap company reviews
As indeed.com has anti scrapping measures, there are times where the code is interuppted and have to be run a few times. To ensure that the code picks up where it has stopped previously, we have: <br>
the no_sg_list, where companies previously visited but have no reviews from singapore being recorded. They will not be visited again.<br>
The additional condition using review_dict["company"]. The condition ensures that we do not revist companies we have already scrapped before.

```python

no_sg_list = []
```

```python
company_namelist = company_names + additional_companies
for company in company_namelist:
    if company not in review_dict["company"] and company not in no_sg_list:
        search_company(company) #search the comapny on the company review search bar
        try: #sometimes, these steps fail for random reasons, if that is the case, the company's review will not be extracted
            industry = click_most_relevant_result() 
            check = check_location()
            if check == True:
                scrap_review_page(company,industry)
            else:
                no_sg_list.append(company)
        except:
            no_sg_list.append(company)
        #return to previous page
        company_review_button = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.CSS_SELECTOR, "[id='CompanyReviews']")))
        company_review_button.click()
    else:
        print(company)
        pass
```
